```python
sum1 = 0
for i in range(10):
  sum1 = sum1 + (i+1)

sum2 = 0
for j in range(30):
  sum2 = sum2 + (j+1)

print(sum1 + sum2)
```

```python
def sum_one2n(n):
  sum = 0
  for i in range(n):
    sum = sum + (i+1)
  return sum

sum1 = sum_one2n(10)
sum2 = sum_one2n(30)

print(sum1 + sum2)
```

```python
def sum_all(x,y,z):
  tmp = x+y
  return (tmp+z)

l = 3
m = 2
n = 5

print(sum_all(l,m,n))
```

```python
def next_term(a):
  return 2*a + 3

a = 1
for _ in range(9):
  a = next_term(a)

print(a)
```

```python
def gcd(n, m):
  while m!=0:
    r = n % m
    n = m
    m = r

  return n
```

```python
def gcd_recursive(n,m):
  if m == 0:
    return n
  else:
    return gcd_recursive(m,n%m)
```

```python
gcd(120,12)
```

```python
gcd_recursive(120,12)
```
