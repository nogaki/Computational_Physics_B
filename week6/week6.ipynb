## 刻み幅を変えながら、数値微分の精度検証

```python
import numpy as np
import matplotlib.pyplot as plt

h_arr = 10.0 ** (-np.arange(0, 15)) # 1から1e-15までの配列を生成
X = 0.3*np.pi # x = 0.3πで微分
```

```python
#　数値微分の実行
derivative = []
for ih in h_arr:
  a = (np.sin(X + ih) - np.sin(X))  / ih
  derivative.append(a)

derivative = np.array(derivative)
```

```python
# 両対数グラフで描画

plt.xlim([1.0,1e-16])
#plt.ylim([1e-10,1.0])
plt.title("Accuracy of numerical derivative of sin(x)", fontsize=14)
plt.xlabel("h", fontsize=20)
plt.ylabel("error", fontsize=20)
plt.tick_params(labelsize=16)
plt.xscale('log')
plt.yscale('log')
plt.plot(h_arr, np.abs(derivative - np.cos(X)),marker='^')
```
