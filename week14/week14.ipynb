## 二分法

```python
def f(x):
  return x**3  - x**2 + x -1
```

```python
def Bisection(f, xmin, xmax):
  assert f(xmin)*f(xmax) < 0

  n_iter = 500
  tol = 1e-3

  for _ in range(n_iter):
    mid = 0.5 * (xmin + xmax)
    f_mid = f(mid)

    if f_mid * f(xmin) < 0:
      xmax = mid
    else:
      xmin = mid

    print(f"xmin: {xmin:.4f}, xmax: {xmax:.4f}")

    if abs(xmin - xmax) < tol:
      print(f"result: {0.5 * (xmin + xmax):.4f}")
      return 0.5 * (xmin + xmax)
```

```python
Bisection(f, -1, 2)
```

## Newton法

```python
def f_deriv(x):
  return 3*x**2  - 2*x + 1
```

```python
def Newton(f, f_deriv, xinit):

  n_iter = 50
  tol = 1e-3

  xprev = xinit

  for _ in range(n_iter):
    xinit = xinit - f(xinit)/f_deriv(xinit)
    print(f"xinit: {xinit:.4f}")

    if abs(xprev - xinit) < tol:
      print(f"result: {xinit:.4f}")
      return xinit

    xprev = xinit
```

```python
Newton(f, f_deriv, 3)
```

## 固定点問題

```python
def g(x):
  return -x**3  + x**2 + 1
```

```python
def Fixed(g, x):
  n_iter = 50
  tol = 1e-3
  xprev = x

  r = 0.5

  for _ in range(n_iter):
    x = r*g(x) + (1-r)*x
    print (f"x: {x:.4f}")

    if abs(x - xprev) < tol:
      print(f"result: {x:.4f}")
      return x

    xprev = x
```

```python
Fixed(g, 1.5)
```
